## Lab | Revisiting Machine Learning Case Study

```python
df = pd.read_csv('../learningSet.csv')
df
```

```python
nulls = pd.DataFrame(df.isna().sum()*100/len(df), columns=['percentage'])
nulls.sort_values('percentage', ascending = False).head(50)
```

```python
drop_list = ['OSOURCE', 'ZIP']

columns_to_drop = nulls[nulls['percentage'] > 85].index.tolist()

drop_list.extend(columns_to_drop)
```

```python
df = df.drop(drop_list , axis=1)
```

```python
print(df['GENDER'].value_counts())
df['GENDER'] = df['GENDER'].fillna('F')
```

```python
df['GENDER'] = df['GENDER'].apply(lambda x: x if x in ['F', 'M'] else 'O')
```

```python
df['GENDER'].value_counts()
```

## Lab | Feature engineering

```python
num_col = df.select_dtypes('number')
```

```python
for i in num_col.columns:
    print(f"{i}: {num_col[i].isna().sum()} missing values")
```

```python
## Using appropriate methods we will now clean the columns GE0C0DE2, WEALTH1, ADI, DMA, MSA
```

```python
#GEOCODE2: For this columns as the NaNs are very small and also we have a small amount of whitespace
#I will convert both to the Mode value. 
print(df['GEOCODE2'].isna().sum())
df['GEOCODE2'].value_counts()
```

```python
mode_value = df['GEOCODE2'].mode().iloc[0]
df['GEOCODE2'].replace(' ',mode_value, inplace = True)
df['GEOCODE2'].fillna(mode_value, inplace = True)
```

```python
df['GEOCODE2'].value_counts()
```

```python
# As the rest of the columns we have left to clean are numerical
#I will start with checking the distribution of these columns
clean_columns = ['WEALTH1','ADI','DMA','MSA']
for col in clean_columns:
    sns.distplot(df[col])
    plt.show()
```

```python
for i in clean_columns:
    print(df[i].isna().sum())
    print(df[i].value_counts())
```

```python
#as wealth has a lot of NaNs we want to keep as many as possible, I googled the issue and found out about imputers
#I used KNNImputer to impute the missing values
imp = KNNImputer(n_neighbors=2, weights = 'uniform')
df['WEALTH1'] = imp.fit_transform(df[['WEALTH1']])
```

```python
df['WEALTH1'].value_counts()
```
